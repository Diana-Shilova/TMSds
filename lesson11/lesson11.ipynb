Операции сложения, вычитания, умножения матриц, первоначально решив на листике (все задачи с данными действиями были решены верно)

```python
AM = np.array([
    [1, 2, 4],
    [7, 5, -2],
    [-2, 1, 3]
])

BM = np.array([
    [-1, 0, 3],
    [2, -6, 1],
    [7, 1, 5]
])

CM = np.array([
    [-2, 3],
    [4, 0],
    [7, -1]
])
```

```python
summ = np.add(AM, BM)
subst = np.subtract(AM, BM)
mult = np.dot(AM, BM)
mult_new = np.dot(mult, CM)

print( f'sum \n {summ}')
print(f'subtract \n {subst}')
print(f'multiply 2 matrix \n {mult}')
print(f'multiply 3 matrix \n {mult_new}')
```

Свертка фотографии

```python
from scipy.signal import convolve2d
import matplotlib.pyplot as plt
```

```python
im = plt.imread("Images/Image1.png")
im.shape
```

```python
def plti(im, h=8, **kwargs):
    """
    Helper function to plot an image.
    """
    
    y = im.shape[0]
    x = im.shape[1]
    w = (y/x) * h
    plt.figure(figsize=(w,h))
    plt.imshow(im, interpolation="none", **kwargs)
    plt.axis('off')

plti(im)
```

Размытие по Гауссу

```python
bw = im.mean(axis = 2)
plt.imshow(bw, cmap = 'gray')
plt.show()
```

```python
#создание двухмерного Гаусового изображения

W = np.zeros((20, 20))

for i in range(20):
    for j in range(20):
        dist = (i - 9.5) ** 2 + (j - 9.5) ** 2
        W[i, j] = np.exp( -dist / 50)

plt.imshow(W, cmap= 'gray')
plt.show()
```

```python
#свертка изображения

out = convolve2d(bw, W)
plt.imshow(out, cmap = 'gray')
plt.show()
```

```python
out.shape
```

```python
#без черных краев на фотографии

out = convolve2d(bw, W, mode = 'same')
plt.imshow(out, cmap = 'gray')
plt.show()
```

Свёртки с распознаванием контура

```python
#оператор Собеля - определяется для двух направлений с последующей аппроксимацией градиента для каждой точки изображения

Hx = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1],
])

Hy = Hx.T
```

```python
Gx = convolve2d(bw, Hx) # G - gradient, gorizont contur
plt.imshow(Gx, cmap = 'gray')
plt.show()
```

```python
Gy = convolve2d(bw, Hy) #vertical contur
plt.imshow(Gx, cmap = 'gray')
plt.show()
```

```python
G = np.sqrt(Gx * Gx + Gy * Gy) # находим контуры
plt.imshow(G, cmap = 'gray')
plt.show()
```

```python
theta = np.arctan2(Gy, Gx) # находим направление градиентов
plt.imshow(theta, cmap = 'gray')
plt.show()
```

```python
def convolve2d(image, kernel):
    """
    This function which takes an image and a kernel and returns the convolution of them.

    :param image: a numpy array of size [imageH, imageW, num_channels].
    :param kernel: a numpy array of size [kernelH, kernelW].
    :return: a convolution output.
    """

    kernelH, kernelW = kernel.shape
    imageH, imageW, num_channels = image.shape

    arr = np.zeros((imageH - kernelH + 1, imageW - kernelW + 1, num_channels)) #размер итоговой матрицы

    for channel in range(num_channels):
        for x in range(imageH - kernelH + 1):
            for y in range(imageW - kernelW + 1):
                arr[x, y, channel] = np.sum(
                                                image[x:x + kernelH, y:y + kernelW, channel] * kernel).clip(0, 255) # значения вне интервала обрезаются

    return arr

kernel_1 = np.array([[-1, -1, -1],
                   [-1, 9, -1],
                   [-1, -1, -1]])              

num_img = 4

fig, ax = plt.subplots(num_img, 2, figsize = (5, 5))
ax[0,0].set_title("Before")
ax[0,1].set_title("After")

for i in range(num_img):
    img = plt.imread(f"Images/Image{i+1}.jpeg")
    convolved_image1 = convolve2d(img, kernel_1)

    ax[i,0].imshow(img)
    ax[i,0].axis('off')

    ax[i,1].imshow(convolved_image1)
    ax[i,1].axis('off')

plt.show()
```
