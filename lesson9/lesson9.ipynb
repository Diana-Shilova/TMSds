```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('dataset/BostonHousing.csv')
df
```

```python
print("statistics for Boston housing:")
print(f"min price: {np.min(df['medv'])}")
print(f"max price: {np.max(df['medv'])}")
print(f"mean price: {np.mean(df['medv'])}")
print(f"median price {np.median(df['medv'])}")
```

```python
print("stat for count of rooms:")
print(f"min rooms: {np.min(df['rm'])}")
print(f"max rooms: {np.max(df['rm'])}")
print(f"mean rooms: {np.mean(df['rm'])}")
```

```python
print("stat for level of poberty:")
print(f"min lstat: {np.min(df['lstat'])}")
print(f"max lstat: {np.max(df['lstat'])}")
print(f"mean lstat: {np.mean(df['lstat'])}")
```

```python
print(f"number of house which located near river:", f"{df['chas'].sum()}")
```

```python
plt.figure(figsize = (25, 15))
n = 1

for column in df:
    if n <= 14:
        ax = plt.subplot(3, 5, n)
        sns.distplot(df[column])
    n += 1

plt.show()
```

```python
plt.figure(figsize = (25, 15))
n = 1

for column in df:
    if n <= 13:
        ax = plt.subplot(3, 5, n)
        plt.scatter(data = df, x = column, y = 'medv')
        plt.xlabel(column) 
        plt.ylabel('Price')
    n += 1

plt.suptitle('Correlation beetween Price and other factor')
plt.show()
```

```python
plt.figure(figsize = (1, 1))
price= sns.lmplot(x= "rm", y= "medv", data= df, col= "chas", order = 2)
price.fig.suptitle("Price and Room corr with Charles River Durmy Variable", fontsize= 16) 
price.fig.subplots_adjust(top= 0.8)
plt.show()
```

```python
plt.figure(figsize = (1, 1))
price= sns.lmplot(x= "nox", y= "medv", data= df, col= "chas", order = 2)
price.fig.suptitle("Price and nitric oxides concentration ", fontsize= 16) 
price.fig.subplots_adjust(top= 0.8)
plt.show()
```

```python
price= sns.lmplot(x= "chas", y= "medv", data= df, x_estimator = np.mean)
price.fig.suptitle("Price and location house near the river or not ", fontsize= 16) 
price.fig.subplots_adjust(top= 0.85)
plt.show()
```

```python
plt.figure(figsize = (1, 1))
price= sns.lmplot(x= "lstat", y= "medv", data= df)
price.fig.suptitle("Price and level of poberty", fontsize= 20) 
price.fig.subplots_adjust(top= 0.8)
plt.show()
```

```python

plt.figure(figsize = (1, 1))
price= sns.lmplot(x= "lstat", y= "rm", data= df)
price.fig.suptitle("Room and level of poberty", fontsize= 20) 
price.fig.subplots_adjust(top= 0.8)
plt.show()
```
