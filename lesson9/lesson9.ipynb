```python
plt.figure(figsize = (25, 15))
n = 1

for column in df:
    if n <= 14:
        ax = plt.subplot(3, 5, n)
        sns.distplot(df[column])
    n += 1

plt.show()
```

```python
plt.figure(figsize = (25, 15))
n = 1

for column in df:
    if n <= 13:
        ax = plt.subplot(3, 5, n)
        plt.scatter(data = df, x = column, y = 'medv')
        plt.xlabel(column) 
        plt.ylabel('Price')
    n += 1

plt.show()
```

```python
plt.figure(figsize = (1, 1))
price= sns.lmplot(x= "rm", y= "medv", data= df, col= "chas", order = 2)
price.fig.suptitle("Price and Room corr with Charles River Durmy Variable", fontsize= 16) 
price.fig.subplots_adjust(top= 0.8)
plt.show()
```

```python
plt.figure(figsize = (1, 1))
price= sns.lmplot(x= "nox", y= "medv", data= df, col= "chas", order = 2)
price.fig.suptitle("Price and nitric oxides concentration ", fontsize= 16) 
price.fig.subplots_adjust(top= 0.8)
plt.show()
```

```python
price= sns.lmplot(x= "chas", y= "medv", data= df, x_estimator = np.mean)
price.fig.suptitle("Price and location house near the river or not ", fontsize= 16) 
price.fig.subplots_adjust(top= 0.85)
plt.show()
```

```python
plt.figure(figsize = (1, 1))
price= sns.lmplot(x= "lstat", y= "medv", data= df)
price.fig.suptitle("Price and level of poberty", fontsize= 16) 
price.fig.subplots_adjust(top= 0.8)
plt.show()
```

Выводы:

1. Небольшое количество домов располагается вблизи реки, что может указывать на низкую плотность застройки

2. Цены на дома вблизи реки выше, чем располагающиеся дома вдали. При чем анализируя цены в зависимости от расположения дома в среднем цена домов у реки стартует от 14 тыс долл США

3. Цены выше в тех местах, где преступность минимальна

4. Стоимость жилья так же напрямую связана с таким фактором как близость расположения торговой недвижимости. Цена начинает снижаться в тех местах, где есть поблизости торговые точки (наличие развитой инфраструктуры - магазины, бизнес-центры и др)

5. В зависимости от того насколько велика концентрация оксида азота - стоимость земли будет снижаться.

6. В ходе исслодвания так же была обнаружена корреляция цены дома и уровнем бедности населения - чем беднее население тем выше риск преступности на районе, в следствии чего страдает покупная способнасть людей и ведет к низкой стоимости жилья
